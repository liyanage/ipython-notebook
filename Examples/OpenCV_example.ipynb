# OpenCV Example
This is a simple example to get started using the OpenCV Python interface. Extract features in a pair of images and find matches.

(This example is based on OpenCV's [find_obj.py](https://github.com/Itseez/opencv/blob/master/samples/python2/find_obj.py) sample script.)

## Setup
First, setup some initial imports.

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
import os
import urllib2
```

Define a function for getting images from urls so that we can get the OpenCV sample images. Also, add a simple function for visualizing an image inline using Matplotlib.

```python
def load_sample_image(filename):
   response = urllib2.urlopen('https://raw.github.com/Itseez/opencv/master/samples/c/' + filename)
   return cv2.imdecode(np.fromstring(response.read(), dtype='uint8'), 0)
```

```python
def show_image(img):
    plt.imshow(img)
    if len(img.shape)<3:
        plt.gray()
    plt.axis("off")
    plt.show()
```

Load the images.

```python
img1 = load_sample_image('box.png')
img2 = load_sample_image('box_in_scene.png')
```

Check what we just loaded.



```python
print("Read img1 of size {} and img2 of size {}.".format(img1.shape, img2.shape))
print("Image 1:")
show_image(img1)
print("Image 2:")
show_image(img2)
```

## Feature matching
Setup a SIFT feature detector and a simple brute force matcher using L2 distance.

```python
detector = cv2.SIFT()
norm = cv2.NORM_L2
matcher = cv2.BFMatcher(norm)
```

Next, extract SIFT keypoints and their descriptors. The optional argument lets you use a mask for where to apply the feature extraction. Set it to None in this case.

```python
kp1, desc1 = detector.detectAndCompute(img1, None)
kp2, desc2 = detector.detectAndCompute(img2, None)
```

Use k-nearest neighbor matching to get the two closest matches (k=2).

```python
raw_matches = matcher.knnMatch(desc1, trainDescriptors = desc2, k = 2)
```

Define a function for cleaning up matches. The idea is to only keep keypoint correspondences where the closest match is significantly better than the second closest point. This idea was introduced by David Lowe and described in the papers in his SIFT [website](http://www.cs.ubc.ca/~lowe/keypoints/).

```python
def filter_matches(kp1, kp2, matches, ratio = 0.75):
    """
    Keep only matches that have distance ratio to 
    second closest point less than 'ratio'.
    """
    mkp1, mkp2 = [], []
    for m in matches:
        if m[0].distance < m[1].distance * ratio:
            m = m[0]
            mkp1.append( kp1[m.queryIdx] )
            mkp2.append( kp2[m.trainIdx] )
    p1 = np.float32([kp.pt for kp in mkp1])
    p2 = np.float32([kp.pt for kp in mkp2])
    kp_pairs = zip(mkp1, mkp2)
    return p1, p2, kp_pairs
```

```python
p1, p2, kp_pairs = filter_matches(kp1, kp2, raw_matches)
```

Take the filtered matches and estimate a homography that transforms the points between the two images using RANSAC.

```python
H, status = cv2.findHomography(p1, p2, cv2.RANSAC, 5.0)
```

This returns the homography and a boolean array indicating which points are inliers. The homography looks like this:

```python
print(H)
```

The inliers of the estimated homography should be high quality matches.

## Visualization
We can plot the inlier matches to see if they are correct.

Create a new image with the original images side-by-side. Convert the grayscale images to OpenCV BGR color images so that we can use the color channels for plotting. 

```python
h1, w1 = img1.shape[:2]
h2, w2 = img2.shape[:2]
vis = np.zeros((max(h1, h2), w1+w2), np.uint8)
vis[:h1, :w1] = img1
vis[:h2, w1:w1+w2] = img2
vis = cv2.cvtColor(vis, cv2.COLOR_GRAY2BGR)
```

Now we can look at this new image.

```python
show_image(vis)
```

_(You can resize the image by dragging at the bottom right corner.)_

Now get the keypoint coordinates and add w1 to the ones in the second image.

```python
p1 = np.int32([kpp[0].pt for kpp in kp_pairs])
p2 = np.int32([kpp[1].pt for kpp in kp_pairs]) + (w1, 0)
```

Plot the matches with green lines.

```python
# plot the matches
color = (0, 255, 0)
for (x1, y1), (x2, y2), inlier in zip(p1, p2, status):
    if inlier:
        cv2.circle(vis, (x1, y1), 2, color, -1)
        cv2.circle(vis, (x2, y2), 2, color, -1)
        cv2.line(vis, (x1, y1), (x2, y2), color)
```

Finally, look at the matches.

```python
show_image(vis)
```
